```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['sl','sw','pl','pw','class']
df=pd.read_csv("iris.csv",header=None,names=cols)
print(df)
```

```python
df.describe()
```

```python
column = df["sl"]
print("Mean =", column.mean())
print("Median =", column.median())
print("Mode =")
print(column.mode())
print("Variance =", column.var())
print("Standard Deviation =", column.std())
```

```python
Q1 = column.quantile(0.25)
Q2 = column.quantile(0.50)
Q3 = column.quantile(0.75)

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
```

```python
IQR = Q3 - Q1
print("IQR =", IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(column < lower) | (column > upper)]

print(outliers)
```

```python
plt.hist(column, bins=10, color="skyblue", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Sl")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(column)
plt.title("Boxplot")
plt.show()
```
