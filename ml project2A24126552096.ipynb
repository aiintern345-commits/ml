# IMPORT

```python

import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
```

# load dataset

```python

df = pd.read_csv(r"C:\Users\anits-csm\Desktop\charancsm\iris.csv")

print("First Five Records")
print(df.head())

print("\nDataset Information")
print(df.info())
```

# DATA CLEANING

# Handling Missing Values

```python

print("\nMissing Values Before Cleaning")
print(df.isnull().sum())

imputer = SimpleImputer(strategy='mean')

numeric_columns = df.select_dtypes(include=np.number).columns

df[numeric_columns] = imputer.fit_transform(df[numeric_columns])

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

# Handling Noisy Data (Binning)


```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\anits-csm\Desktop\charancsm\iris.csv",
    header=None,
    names=[
        "SepalLength",
        "SepalWidth",
        "PetalLength",
        "PetalWidth",
        "Species"
    ]
)

print(df.head())
print(df.columns)
```

```python
df["SepalLength_Bin"] = pd.cut(
    df["SepalLength"],
    bins=3,
    labels=["Short", "Medium", "Long"]
)

print(df[["SepalLength", "SepalLength_Bin"]].head())
```

#  Handling Outliers using IQR

```python
Q1 = df["SepalWidth"].quantile(0.25)
Q3 = df["SepalWidth"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df["SepalWidth"] < Q1 - 1.5*IQR) |
    (df["SepalWidth"] > Q3 + 1.5*IQR)
]

print("\nOutliers")
print(outliers)
```

#  DATA INTEGRATION

```python

df2 = df[
    ["Species","PetalLength"]
]

merged_data = pd.merge(
    df,
    df2,
    on="Species"
)

print("\nIntegrated Dataset")
print(merged_data.head())
```


# Removing duplicate records


```python
print("\nDuplicate Records:")
print(df.duplicated().sum())

df = df.drop_duplicates()
```

# DATA TRANSFORMATION

```python
features = [
    "SepalLength",
    "SepalWidth",
    "PetalLength",
    "PetalWidth"
]
```

#  Normalization

```python



scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(
    df[features]
)

normalized_df = pd.DataFrame(
    normalized_data,
    columns=features
)

print("\nNormalized Data")
print(normalized_df.head())
```

#  Standardization

```python

standard = StandardScaler()

standardized_data = standard.fit_transform(
    df[features]
)

standardized_df = pd.DataFrame(
    standardized_data,
    columns=features
)

print("\nStandardized Data")
print(standardized_df.head())
```

# Aggregation

```python

aggregation = df.groupby("Species").mean(
    numeric_only=True
)

print("\nAggregated Data")
print(aggregation)
```

# Discretization

```python

df = df.copy()

df.loc[:, "PetalLength_Category"] = pd.cut(
    df["PetalLength"],
    bins=3,
    labels=[
        "Small",
        "Medium",
        "Large"
    ]
)

print("\nDiscretization")
print(df[["PetalLength", "PetalLength_Category"]].head())
```

#  Encoding Categorical Data

```python

encoder = LabelEncoder()

df["Species_Label"] = encoder.fit_transform(
    df["Species"]
)

print("\nLabel Encoding")
print(df[["Species","Species_Label"]].head())

```

# One Hot Encoding

```python


one_hot = pd.get_dummies(
    df["Species"],
    prefix="Species"
)

df = pd.concat(
    [df,one_hot],
    axis=1
)

print("\nOne Hot Encoding")
print(df.head())
```

#  DATA REDUCTION

```python


X = df[features]

pca = PCA(
    n_components=2
)

pca_data = pca.fit_transform(X)


pca_df = pd.DataFrame(
    pca_data,
    columns=[
        "PC1",
        "PC2"
    ]
)

print("\nPCA Result")
print(pca_df.head())
```

# Feature Selection

```python

X = df[features]

y = df["Species_Label"]

selector = SelectKBest(
    score_func=f_classif,
    k=2
)

selected_features = selector.fit_transform(
    X,
    y
)

print("\nSelected Features")
print(selected_features[:5])
```

#  Numerosity Reduction - Sampling

```python


sample_data = df.sample(
    n=20,
    random_state=1
)

print("\nSample Data")
print(sample_data)
```

# FINAL DATASET

```python

print("\nFinal Preprocessed Dataset")
print(df.head())

print("\nPreprocessing Completed Successfully")
```
